```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

alpha = 0.05

from scipy.stats import ttest_ind, ttest_1samp, ttest_rel
```

```python
df = pd.read_csv('./data/qwantify.csv')
```

```python
df.head()
```

```python
a = df.ix[df['wantingAnything']==2, 'appreciating'].values
b = df.ix[df['wantingAnything']==0, 'appreciating'].values

stat, p = ttest_ind(a, b, equal_var=False)

print 'Average appreciating value when having a want:{}'.format(round(np.mean(a), 3))
print 'Average appreciating value when NOT having a want:{}'.format(round(np.mean(b), 3))

print '\nTest Statistics t = {}'.format(round(stat,2))
print 'Two tailed independent t-test p = {}'.format(round(p, 2))
print 'One tailed independent t-test p = {}'.format(round(p/2, 2))

if p/2 < alpha:
    print '\nSignificant results for one-sided t-test.'
elif p < alpha:
    print '\nSignificant results for two-sided t-test.'
else:
    '\nTwo samples are not significantly different.'
```
