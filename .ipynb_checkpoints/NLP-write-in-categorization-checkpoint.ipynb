I think to start with I'll just implement a basic algorithm that goes through every word in the document, and compares it's corresponding word vector with a list of words (vectors) that I want to generalize to. For this example, my "master word list" will be the "doing" verbs in the qwantify app. These verbs are:
<br><br>
`['talk', 'chores', 'other', 'work', 'eat', 'travel', 'leisure','sleep', 'exercise']`

I will also add the following verbs:<br>
1. socialize 
2. relax


**Get words to compare**

```python
wanting = [want for want in df['whatWanting'].unique().tolist() if str(want)!='nan']

doing = df['doing'].unique().tolist()
doing.append('socialize')
doing.append('relax')
```

**Pre-process words**


```python
stops = set(stopwords.words("english"))

categorizations = []
for want in wanting: # for every user-input phrase in the whatWanting field

    want = want.translate(None, string.punctuation) # remove punctuation 
    
    words = want.lower().split() 
    
    words = [nlp(w.decode('UTF-8')).text for w in words if not w in stops] # remove stop words  
    
    matches = []
    for i, word in enumerate(words): # for word in phrase
        
        word = nlp(word.decode('UTF-8')) # convert to nlp object
        
        sim = 0
        max_sim = 0
        for verb in doing: # compare each word to every word in the doing list (master list of verbs)
            
            sim = word.similarity(nlp(verb.decode('UTF-8')))
            
            if sim > max_sim:
                max_sim = sim
                best_match = verb
                print word, best_match, sim
        
        matches.append([word, best_match, max_sim]) 
#         matches.append([best_match, max_sim]) 
        
    best_match = sorted(matches, key=lambda x: x[2], reverse=True)[0]
#     best_match = sorted(matches, key=lambda x: x[1])[0]
    categorizations.append(([want] + best_match))
```

Problem with this method is that:
1. it only looks at one word-verb pair, and that is the pair with the highest similarity - all other pairs are ignored in the final category classification. 
2. This implementation isn't specific to emotions and words. We could fine tune the algorithm so that, if we were comparing the phrases only to verbs, then we could remove all non-verbs from each phrase. This would solve problems like 'wash my car' being tagged as 'travel' because 'car' matches 'travel' most. But it would not work well on 'Make dinner' which would probably get tagged as '

### Algorithm 2


```python
stops = set(stopwords.words("english"))

categorizations = []
for j, want in enumerate(wanting): # for every user-input phrase in the whatWanting field

    want = want.translate(None, string.punctuation) # remove punctuation 
    
    words = want.lower().split() 
    
    words = [nlp(w.decode('UTF-8')).text for w in words if not w in stops] # remove stop words  
    
    matches = []
    
    # Initialize dictionary for counting phrase-verb match frequency 
    d = dict()    
    for verb in doing:
        d[verb] = {
                'count':0,
                'sim_scores':[] # sim_scores is only used in the event of a frequency tie between two verbs
                                # use the verb with the highest simiarilty scores
        }
        
    for i, word in enumerate(words): # for word in phrase
        
        word = nlp(word.decode('UTF-8')) # convert to nlp object
        
        sim = 0
        max_sim = 0
        for verb in doing: # compare each word to every word in the doing list (master list of verbs)
            
            sim = word.similarity(nlp(verb.decode('UTF-8')))
            
            if sim > max_sim:
                max_sim = sim
                best_match = verb
#                 print word, best_match, sim
        
        # Update match-frequency dict 
        d[best_match]['count'] += 1
        d[best_match]['sim_scores'].append(sim)
        
    
    # sort dictionary by counts
    ordered = OrderedDict(sorted(d.items(), key=lambda x: x[1]['count'], reverse=True))
#     pprint(d)
    
    # find greatest match frequency
    top_score = np.max(ordered.items()[0][1]['sim_scores'])
    top_verb = ordered.items()[0][0]
    max_count = ordered.items()[0][1]['count']
    for match in ordered.items():
        verb = match[0]
        count = match[1]['count']
        if count == max_count: # if there is a tie in frequency
            max_verb_score = np.max(match[1]['sim_scores']) # take the maximum match score
            if max_verb_score > top_score: # compare that with 
                top_verb = verb 
                top_score = max_verb_score

    
    categorizations.append(([j] + [want] + [top_verb, top_score]))
```
